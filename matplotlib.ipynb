```python
import matplotlib.pyplot as plt
import numpy as np
```

# **Scatter plot**

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.scatter(x,y, c='b',marker="^",alpha=0.6)
plt.show()
```

```python
plt.title('Basic Matplotlib Example of scattery plot')
x = np.random.rand(100)
y = np.random.rand(100)
plt.xlabel("X-AXIS")
plt.ylabel("Y-Axis")
plt.scatter(x,y, c='b',marker="^",alpha=0.6)
plt.show()
```

```python
x_new= [5,6,7,8]
y_new = [99,75,42,100]
# sb ky alag alag marker sizes
sizes = [100,20,7,77]
colors=[1,2,3,4]
plt.scatter(x_new,y_new,s=sizes, c=colors ,cmap='viridis',alpha=0.7)
plt.title("Scatter plot example")
plt.xlabel("X-AXIS")
plt.ylabel("Y-Axis")
plt.colorbar(label='color scale')
plt.show()
```

# plot with color marker line style and line width and label

```python
years= [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009]
weight=[80,75,70,65,60,55,50,45,40,35]
#lw means line width marker can be anything ^,*
plt.plot(years, weight,c='#B565A7',lw=4,marker="o",linestyle="--")
plt.title("Line plot example")
plt.xlabel("Years")
plt.ylabel("Weight")
plt.show()
```

# grid aur label lagana deko label='2x line'

```python
x = np.random.rand(10)
y = np.random.rand(10)
plt.plot(x,y,color='#FF8C00',marker="o", linestyle="--", linewidth=2, label='2x line')
plt.title("Line plot example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

# **Seaborn**

```python
import seaborn as sns
dataset_names= sns.get_dataset_names()
print(dataset_names)
```

```python
data = sns.load_dataset("penguins")
print(data)
```

```python
data.head()
```

```python
sns.countplot(data=data, x='island', hue='sex',palette='Set2',edgecolor='black')
plt.legend(title='Gender', loc='upper right')
plt.title("Penguins Gender Distribution by Island")
plt.xlabel("Island")
plt.ylabel("Count")
plt.show()
# palete toh colors hein like set2,mako,flare
```

```python
avg_mass = data.groupby('species')['body_mass_g'].mean().reset_index()

# Line plot
sns.lineplot(data=avg_mass, x='species', y='body_mass_g', marker='o', color='teal')

plt.title("Average Body Mass per Penguin Species")
plt.xlabel("Species")
plt.ylabel("Average Body Mass (g)")
plt.show()
```

```python
sns.scatterplot(
    data=data,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    style='sex',
    palette='viridis',
    alpha=0.8
)

plt.title("Flipper Length vs Body Mass (by Species & Gender)", fontsize=14)
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()
```
